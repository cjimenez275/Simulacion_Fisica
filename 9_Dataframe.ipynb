# Creación y manipulación de datos

```python
import pandas as pd
```

## 1. Creación de un DataFrame

Podemos crear un DataFrame de múltiples maneras, una de ellas es por medio de un diccionario

```python
encabezado = ["Reproducción", "Volumen", "Género"]
datos = {"Por que te fuiste": [90, 10, "Tango"],
         "Mala gente": [15, 8, "Pop"],
         "Eres": [6, 8, "Rock"],
         "Adios": [54, 9, "Tango"],
         "Mi vida": [86, 5, "Pop"]}

musica = pd.DataFrame(datos, index=encabezado)
musica = musica.T
musica
```

Seleccionar una columna

```python
musica["Volumen"]
```

```python
musica["Reproducción"]
```

De una columna seleccionar un elemento

```python
musica["Reproducción"].iloc[2]
```

```python
musica.iloc[2]
```

Agregar una línea entera

```python
datos2 = {"Por tu amor": [80, 46, "Tango"]}

musica2 = pd.DataFrame(datos2, index=encabezado)
musica2 = musica2.T
musica2
```

```python
musica = pd.concat([musica, musica2])
musica
```

Podemos agregar una columna nueva

```python
likes = [5, -2, 4, 12, 0, 8]
musica["Likes"] = likes
musica
```

Podemos eliminar filas por el nombre de la fila o por el índice de la fila, incluso eliminar varias con
el comando $[i:j]$

```python
musica.drop(musica.index[1:2], inplace=True)
```

```python
musica
```

```python
musica.drop(["Adios"], axis=0, inplace=True)
musica
```
