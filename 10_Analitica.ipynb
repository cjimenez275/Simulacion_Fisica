# Gr√°ficos con SeaBorn

```python
sns.get_dataset_names();
```

```python
pinguinos = sns.load_dataset("penguins")
```

```python
pinguinos.columns = ["especie", "isla", "longitud_pico",
                     "profundidad_pico", "longitud_aleta", "masa", "sexo"]
pinguinos.head(2)
```

```python
pinguinos.describe()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta", height=7)
plt.show()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="masa")
plt.show()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie")
plt.show()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="sexo")
plt.show()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie",
            row="sexo",
            size="masa",
            sizes=(100, 500),
           style="sexo")
plt.show()
```

```python
sns.violinplot(data=pinguinos, x="longitud_aleta", y="isla")
plt.show()
```

```python
sns.catplot(data=pinguinos, x="isla", y="longitud_aleta", kind='strip', hue='sexo')
plt.show()
```

```python
sns.displot(data=pinguinos, x="longitud_pico", hue="especie", multiple="stack")
plt.show()
```
