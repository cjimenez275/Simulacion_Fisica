## Experimentos

Muestra aleatoria de elementos que tienen correlación

```python
L = []
N = 100
a, b = 100, 20
for i in range(N):
  L.append(int(random.gauss(a, b)))

plt.hist(L)
plt.show()
```

lanzamiento de una moneda

```python
L = ["cara", "sello"]
v = []
N = 10
for i in range(N):
  x = random.choice(L)
  v.append(x)
plt.hist(v)
plt.show()
print("Cara: ", v.count("cara"))
print("Sello: ", v.count("sello"))
print("Probabilidad de cara: ", round(v.count("cara")/N, 2))
print("Probabilidad de sello: ", round(v.count("sello")/N, 2))
```

lanzamiento de dados:
Se lanzan dos dados y se observa el resultado

```python
L = []
N = 10
for i in range(N):
  x = random.randint(1, 6)
  y = random.randint(1, 6)
  R = x + y
  L.append(R)
plt.hist(L, bins=10, rwidth=0.8)
plt.grid(True)
plt.show()
```

## Máquina de monedas

```python
coins = 10
g = 0
for i in range(coins):
  n = random.randint(0, 9)
  m = random.randint(0, 9)
  p = random.randint(0, 9)
  if n == m and m == p:
    g = g + 50
    print(n, m, p, "Ganaste", g, "monedas")

print("Gastaste", coins, "Ganaste", g)
print("Tu saldo es: ", g - coins)
```

Bolsa de pelotas:
Se tiene una bolsa con 9 pelotas rojas y 5 verdes:
Simulamos el ejercicio de sacar dos pelotas.

```python
N = 1000
result = {"rr": 0, "rg": 0, "gr": 0, "gg":0}

for i in range(N):
    ll = ["r", "r", "r", "r", "r", "r", "r", "r", "r", "g", "g", "g", "g", "g"]
    e1 = ll.pop(random.randint(0, len(ll) - 1))
    e2 = ll.pop(random.randint(0, len(ll) - 1))
    if e1 == "r" and e2 == "r":
        result["rr"] = result["rr"] + 1
    if e1 == "r" and e2 == "g":
        result["rg"] = result["rg"] + 1
    if e1 == "g" and e2 == "r":
        result["gr"] = result["gr"] + 1
    if e1 == "g" and e2 == "g":
        result["gg"] = result["gg"] + 1

print("La probabilidad de sacar dos bolas verdes es: ", result["gg"]/N)
print("El valor exacto es: ", round((5/14)*(4/13), 4))
#print(result)

print("La probabilidad de sacar una roja y una verde es: ", (result["rg"]+result["gr"])/N)
print("El valor exacto es: ", round(2*(9/14)*(5/13), 4))

print("La probabilidad de sacar dos bolas rojas es: ", result["rr"]/N)
print("El valor exacto es: ", round((9/14)*(8/13), 4))
print(result)


Balls = list(result.keys())
values = list(result.values())

plt.bar(Balls, values, width = 0.4)
plt.xlabel("Resultados posibles")
plt.ylabel("Resultados obtenidos")
plt.title("Probabilidad condicionada")
plt.show()
```
