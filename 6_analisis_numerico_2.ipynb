# Análisis numérico 2: Diferencias finitas

Ejemplo:
Solucione la ecuación diferencial mediante el método de diferencias finitas:
$$\frac{dy}{dx}=2y\hspace{1cm}y(0)=3$$

```python
import matplotlib.pyplot as plt
import numpy as np
a, b = 0, 2
N = 100
d = (b-a)/N
x = [a]
y0 = 3
y = [y0]

for i in range(N):
    x.append(x[i]+d)
    y.append((2*d + 1)*y[i])


f = lambda z: 3*np.exp(2*z)

plt.plot(x, y, label="Numérica")
plt.plot(x, f(np.array(x)), label="Exacta")
plt.legend()
plt.show()
```
