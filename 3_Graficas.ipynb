## Linspace

```python
x = np.linspace(0,10, 11)
print(x)
print(len(x))
```

```python
print(x**2)
```

```python
x = np.linspace(-10,10, 11)
print(x)
```

```python
z = np.array(range(0,10))
print(z)
```

```python
z = np.array(range(0,100, 7))
print(z)
print(len(z))
```

```python
z = np.array(range(100,0, -8))
print(z)
print(len(z))
```

```python
print(np.sin((z/z**2)))
```

## Funcion lambda

```python
y = lambda x: x**2 - 3*x
```

```python
z = lambda x, y: np.sqrt(x) + 2**y
```

```python
y(3), z(4, 9)
```

## Gráficas

### Graficas de dispersión
Son gráficos de dos variables

```python
# Gráficas con listas
x = [3, 7, 19, 21]
y = [12, 17, 35, 45]
plt.plot(x, y)
plt.show()
```

```python
# Gráficos con linspace
x = np.linspace(-10, 10, 100)
f = lambda x: x**2 - 10
y = f(x)
plt.plot(x, y)
plt.show()
```

```python
# Gráficos con arrays
x = np.array(range(-10,10, 2))
f = lambda x: x**2 - 10
y = f(x)
plt.plot(x, y)
plt.show()
```

## Ejemplo
Se han monitoreado las ventas de postres "las delicias de la konrad" en tres ciudades a lo largo de todo el año.

```python
t = np.linspace(1, 12, 12)
B = [20, 10, 12, 15, 17, 20, 22, 34, 36, 70, 80, 90]
M = [12, 24, 46, 54, 65, 54, 43, 35, 36, 37, 23, 12]
C = [5, 9, 6, 5, 15, 24, 33, 45, 26, 22, 23, 19]
fig = plt.figure(figsize=(10, 3))
plt.plot(t, B, label="Bogotá", marker="v", c="g")
plt.plot(t, M, label="Medellín", marker="*", c="r")
plt.plot(t, C, label="Cali", marker="o", c="b")
plt.title("Las delicias de la K")
plt.legend()
plt.grid()
plt.show()
```
